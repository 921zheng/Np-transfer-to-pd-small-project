```python
import pandas as pd
import numpy as np
```

```python
# create random sale_amount for three different butters
sale_amount=np.random.seed(0)
sale_amount=np.random.randint(20,size=(5,3))
sale_amount
```

```python
#create weekly sale DataFrame
sale_amount_pd=pd.DataFrame(sale_amount,index=["Mon","Tues","Wed","Thurs","Fri"],columns=["Almond Butter","Peanut Butter","Cashewnut butter"])
sale_amount_pd
```

```python
#create price
price=np.array([10,8,12])
price
```

```python
price_pd=pd.DataFrame(price.reshape(1,3),index=["Price"],columns=["Almond Butter","Peanut Butter","Cashewnut butter"])
price_pd
```

```python
#count total_price
total_price=sale_amount_pd.dot(price_pd.T)
total_price
```

```python
#combine total_price with sale_amount_pd
sale_amount_pd["Total ($)"]=total_price
sale_amount_pd
```
